```python
import numpy as np
import PythonTools as PT
from matplotlib import pyplot as plt

%matplotlib inline
```

# A deomsntartion of Principal Componenet Analysis (PCA)

## Generate bivaraite Normally Distributed observations

```python
mu=[0.0, 0.0]
cov=[[1.0, 0.75], [0.75, 1.0]]
Y=np.random.multivariate_normal(mu,cov,10000)
```

## Checking the covariance

```python
C = 1/len(Y) * np.dot(Y.T,Y)
print(C)
```

```python
PT.BV_Plot(X=Y[:,0],Y=Y[:,1], figsize=(6, 6))
```

# Step 1) PCA

```python
D, V = np.linalg.eig (C)
print(D)
```

```python
X = np.dot(Y,V)
C_X = 1/len(X) * np.dot(X.T,X)
print(C_X)
```

```python
fig,ax=plt.subplots(figsize=(6, 6))
PT.BV_Plot(X=X[:,0],Y=X[:,1], figsize=(6, 6),ax=ax)
ax.set_aspect('equal', 'datalim')
```

# Step 2) Sphering

Standardized by standard deviation of each principal component (i.e., the corresponding eigen value)

```python
S=[[D[0]**(-0.50), 0], [0, D[1]**(-0.50)]]
Z=np.dot(X,S)
```

```python
fig,ax=plt.subplots(figsize=(6, 6))
PT.BV_Plot(X=Z[:,0],Y=Z[:,1], figsize=(6, 6),ax=ax)
ax.set_aspect('equal', 'datalim')
```

# Step 3) Sphere-R

After sphering you can rotate back your data to near original coordinate to minimize the mixing between variables
Note: After s

```python
Q = np.dot(Z,V.T)
```

```python
fig,ax=plt.subplots(figsize=(6, 6))
PT.BV_Plot(X=Q[:,0],Y=Q[:,1], figsize=(6, 6),ax=ax)
ax.set_aspect('equal', 'datalim')
```
