# Distance function as a metric

Note that a distance function d: $\chi \times \chi \rightarrow R$ is a **metric** if it satisfies the following properties:

- $d(x,y) \ge 0$ (non-negativity)
- $d(x,y) = 0$ if and only if X=Y
- $d(x,y) = d(Y,x)$ (Symmetry)
- $d(x,z) \ge d(x,y) + d(y,z)$ (triangle inequality)

The last property (triangle inequality) ensures that a distance matrix is positive definite. Any P norm is a metric and can be used in different applications.

# P norm distances

P norm or Minkowski distances are used in mathematics and statistics to measure similarity of data records or distributions. Here we have a simple visualization of these points.

$X = (x_1, x_2, x_3, ..., x_n) \;\;and\;\;  Y = (y_1, y_2, y_3, ..., y_n)$

$D(X,Y) = (\sum | x_i - y_i |^{p})^{\frac{1}{p}}$ 


```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-1,1,50)

power_values = [-2, -1, 0, 1, 2, 3, 4, 50]

fig, axes = plt.subplots(1, len(power_values), figsize = (4*len(power_values),4))

for i, value in enumerate(power_values):
    p = 2**value
    
    y = (1 - abs(x)**p)**(1/p)
    y_n = -y
    
    axes[i].plot(x,y,'b', lw=2)
    axes[i].plot(x,y_n, 'b', lw=2)
    axes[i].set_xlim(-1.2,1.2)
    axes[i].set_ylim(-1.2,1.2)
    axes[i].set_title(r'$p = 2^{%i}$'%(value))
    axes[i].grid()
```

Note that the large values of P represent p = $\infty$ where the maximum coordinate/dimension dictates the distance. P = 2 is the Euclidean distance. 

The P norm distance function is a metric
